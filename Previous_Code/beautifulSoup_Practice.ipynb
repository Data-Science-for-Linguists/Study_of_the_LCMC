```python
#BeautifulSoup4
#Following Video Tutorials at https://www.youtube.com/watch?v=aIPqt-OdmS0
import bs4 as bs
import urllib.request

#sauce = urllib.request.urlopen('https://pythonprogramming.net/parsememcparseface/').read()
sauce = urllib.request.urlopen('http://news.xinhuanet.com/english/2017-10/12/c_136675044.htm').read()
soup = bs.BeautifulSoup(sauce,'lxml')

print(soup)
```

```python
print(soup.title)
```

```python
print(soup.title.string)
```

```python
print(soup.p)
```

```python
print(soup.find_all('p'))
```

```python
for paragraph in soup.find_all('p'):
    print(paragraph.text)
```

```python
print(soup.get_text())
```

```python
for url in soup.find_all('a'):
    print(url.get('href'))
```

```python
#Navigation
nav = soup.nav
print(nav)
```

```python
for url in nav.find_all('a'):
    print(url.get('href'))
```

```python
body = soup.body
for paragraph in body.find_all('p'):
    print(paragraph.text)
```

```python
for div in soup.find_all('div', class_='body'):
    print(div.text)
```

```python
#Tables
table = soup.find('table')
print(table)
```

```python
table_rows = table.find_all('tr')

for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    print(row)
```
